```python
df=pd.DataFrame()
```

```python
from pathlib import Path
```

```python
df['labels']=text_series.map(lambda x : x.split("\\")[-2])
```

```python
df['text']=text_series.map(lambda x : Path(x).read_text())
```

```python
df
```

```python

##Converting docs to vector
curr_idx=0
word2idx={}
tokenized_docs=[]
for doc in df['text']:
    tokens=word_tokenize(doc.lower())
    doc_as_int=[]
    for token in tokens:
        if token not in word2idx:
            word2idx[token]=curr_idx
            curr_idx+=1
        #saving them for later
        doc_as_int.append(word2idx[token])
    tokenized_docs.append(doc_as_int)
```

```python
print(doc_as_int in tokenized_docs)
```

```python
##reverse mapping
idx2word = [k for k , v in word2idx.items()]
```

```python
word2idx
```

```python
idx2word
```

```python
N=len(df['text'])
```

```python
V=len(word2idx) ##unique words
```

```python
tf = np.zeros((N, V))
```

```python
#populate term-frequency count
for i , doc_as_int in enumerate(tokenized_docs):
    for j in doc_as_int:
        tf[i , j]+=1
```

```python
doc_freq = np.sum(tf>0 , axis=0) #doc freq (shape=(V,))
idf = np.log(N/doc_freq)
```

```python
tf.shape
```

```python
idf.shape
```

```python
tf_idf = tf*idf
```

```python
np.random.seed(123)
```

```python
i=np.random.choice(N)
row = df.iloc[i]
print('Lables: ' , row['labels'])
print('Text :' , row['text'].split("\n" , 1)[0])
print("top 10:")
scores=tf_idf[i]
indices=(-scores).argsort()
for j in indices[:10]:
    print(idx2word[j])
```
