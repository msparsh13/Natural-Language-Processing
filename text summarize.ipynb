```python
def summary(text , factor=0.15):
    token = nltk.sent_tokenize(text)
    x = featurizer.fit_transform(token)
   
    s = cosine_similarity(x)
    s /=s.sum(axis=1 , keepdims=True)
    U = np.ones_like(s)/len(s)
    s= (1-factor)*s+factor*U
    eigenvals , eigenvec = np.linalg.eig(s.T)
    scores = eigenvec[: , 0]/eigenvec[: , 0].sum()
    sort_idx = np.argsort(-scores)


     
    for i in sort_idx[:5]:
        print(token[i])
```
