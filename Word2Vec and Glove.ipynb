```python
##Glove
import numpy as np
```

```python
def load_glove_vectors(glove_file):
    with open(glove_file, 'r', encoding="utf-8") as file:
        # unique words
        words = set()
        word_to_vec = {}
        # each line starts with a word then the values for the different features
        for line in file:
            line = line.strip().split()
            # take the word 
            curr_word = line[0]
            words.add(curr_word)
            # rest of the features for the word
            word_to_vec[curr_word] = np.array(line[1:], dtype=np.float64)
            
    return words, word_to_vec
```

```python
def find_cosine_similarity(u, v):
    distance = 0.0
    
    # find the dot product between u and v 
    dot = np.dot(u,v)
    # find the L2 norm of u 
    norm_u = np.sqrt(np.sum(u**2))
    # Compute the L2 norm of v
    norm_v = np.sqrt(np.sum(v**2))
    # Compute the cosine similarity
    cosine_sim = dot/(norm_u*norm_v)
    
    return cosine_sim
```

```python
def word_analogies(a , b , c): ##a-b=c+d
    m_cs=-999
    f_w=""
    arr=word_to_vec[a.lower()]-word_to_vec[b.lower()]+word_to_vec[c.lower()]
    for word in words:
        if word.lower() != a and word.lower() != b and word.lower() != c :
            w=word_to_vec[word]
            cs=find_cosine_similarity(w , arr)
            if cs>m_cs:
                m_cs=cs
                f_w=word
    return f_w , m_cs
```

```python
def most_similar(a):
    arr=word_to_vec[a.lower()]
    final_list=[] 
    for word in words:
        w=word_to_vec[word]
        cs=find_cosine_similarity(w , arr)
        final_list.append([word , cs])
    final_list.sort(reverse=True , key=lambda x: x[1])
    return final_list[1:11]
```
