```python
import pandas as pd
import seaborn as sns
import numpy as np
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Dense , Input
from tensorflow.keras.models import Model
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.optimizers import Adam
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.metrics import roc_auc_score , f1_score , confusion_matrix
```

```python
np.random.seed(0)
tf.random.set_seed(0)
```

```python
df = pd.read_csv(r"C:\Users\Sparsh Mahajan\Downloads\archive (2)\Tweets.csv")
```

```python
df.head()
```

```python
df = df[['text' ,'airline_sentiment']]
```

```python
df
```

```python
df['airline_sentiment'].hist()
```

```python
df = df[df['airline_sentiment']!="neutral"].copy()
```

```python
target_map = { "positive" : 1 , "negative" : 0}
```

```python
df['target'] = df['airline_sentiment'].map(target_map)
```

```python
df.head()
```

```python
train , test  = train_test_split(df , random_state=42)
```

```python
train
```

```python
vectorizer = TfidfVectorizer(max_features=20000)
```

```python
X_train = vectorizer.fit_transform(train['text'])
```

```python
X_test =  vectorizer.transform(test['text'])
```

```python
X_train= X_train.toarray()
X_test= X_test.toarray()
```

```python
X_test.shape
```

```python
Y_train = train['target']
Y_test =  test['target']
```

```python
D = X_train.shape[1]
```

```python
i = Input(shape=(D,))
x= Dense(1)(i)
```

```python
model = Model(i , x)
```

```python
model.summary()
```

```python
model.compile(
    loss= BinaryCrossentropy(from_logits=True) ,
    optimizer = Adam(learning_rate=0.005) ,
    metrics =["accuracy"]
)
    
```

```python
x = model.fit(X_train , Y_train  , 
          validation_data=(X_test , Y_test) ,
          epochs=400 , 
          batch_size=128)
```

```python
import matplotlib.pyplot as plt
```

```python
r=x
```

```python
plt.plot(r.history['loss'] , label="train_loss")
plt.plot(r.history['val_loss'] , label="test_loss")
plt.legend()
```

```python
plt.plot(r.history['accuracy'] , label="train_acc")
plt.plot(r.history['val_accuracy'] , label="test_acc")
plt.legend()
```

```python
P_train = ((model.predict(X_train)>0)*1).flatten()
P_test = ((model.predict(X_test)>0)*1).flatten()
```

```python
cm = confusion_matrix(Y_train , P_train , normalize='true')
```

```python
cm
```

```python
classes=["pos" , "neg"]
df = pd.DataFrame(cm , index=classes , columns=classes)
sns.heatmap(df,fmt='g' );
```

```python
cm = confusion_matrix(Y_test , P_test, normalize='true')
classes=["pos" , "neg"]
df = pd.DataFrame(cm , index=classes , columns=classes)
sns.heatmap(df,fmt='g' );
```

```python
cm
```

```python
Pr_train = model.predict(X_train)
Pr_test = model.predict(X_test)

```

```python
print(" Train roc score" ,  roc_auc_score( Y_train , Pr_train ))
print(" Test roc score" ,  roc_auc_score( Y_test , Pr_test ))
```

```python
print(" Train f1 score" ,  f1_score( Y_train , P_train ))
print(" Test f1 score" ,  f1_score( Y_test , P_test ))
```
